# Build a custom tool to multiply

```python
from langchain_core.tools import tool
```

```python
# Build the tools by giving it
# a. personalized function name in below case multiply_numbers_ravi

@tool  # b. decorator to define a tool so that it can be called from llm
def multiply_numbers_ravi(a: int, b: int) -> int: # c. type hinting to be explicit what data type it takes as input and return as output
    """Multiplies two numbers.""" # d. doc string to make llm understand what the function does
    return a * b
```

```python
# Now we can invoke the tool multiply_numbers_ravi by passing parameter
multiply_numbers_ravi.invoke({"a":2, "b":3})
```

```python
print(multiply_numbers_ravi)
```

```python
print(multiply_numbers_ravi.description)
```

```python
print(multiply_numbers_ravi.args_schema.model_json_schema())
```

```python
print(multiply_numbers_ravi.args)
```

# Structured Tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyNumbersInput(BaseModel):
    a: int = Field(required= True, description="The first number to add")
    b: int = Field(required = True, description="The second number to add")
    c: int = Field(required = False, description="Output of a and b")
```

```python
def multiply_numbers_structured_ravi(a:int, b:int, c: int) -> int:
  c= a * b
  return c
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_numbers_structured_ravi,
    name = "multiply_numbers_structured_ravi",
    description = "Multiplies two numbers",
    args_schema = MultiplyNumbersInput
    )
```

```python
result = multiply_tool.invoke({"a":4, "b":9, "c": 2})
print(result)
```

```python
print(multiply_tool.description)
```

```python
print(multiply_tool.args)
```

# **Base Tool**

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# arg Schema using Pydantic
class MultiplyNumbersInput_new(BaseModel):
    a: int = Field(required= True, description="The first number to add")
    b: int = Field(required = True, description="The second number to add")
    c: int = Field(required = False, description="Output of a and b")
```

```python
class MultiplyTool_New(BaseTool):
  name: str = "multiply_numbers_ravi"
  description: str = "Multiplies two numbers"
  args_schema: Type[BaseModel] = MultiplyNumbersInput_new
  def _run(self, a: int, b: int, c: int) -> int:
    c = a * b
    return c
```

```python
multiply_tool_new= MultiplyTool_New()
result = multiply_tool_new.invoke({"a":4, "b":9, "c": 2})
print(result)
```
