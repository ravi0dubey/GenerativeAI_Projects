```python
from langchain.agents import create_agent
```

```python
agent = create_agent(
    model = llm,
    tools = [search_tool],
)
```

```python
query ={"messages": [("human", "Who is the primeminister of the Canada?")]}
```

```python
result = agent.invoke(query)
result
```
